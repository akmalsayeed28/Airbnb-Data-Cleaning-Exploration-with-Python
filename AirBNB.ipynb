```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb
```

```python
df = pd.read_csv('AIR BNBdatasets.csv')
```

```python
#dropping all the rows which have null values
df.dropna(inplace=True)
```

```python
#checking for null in each columns
df.isnull().sum()
```

```python
#checking for duplicate rows and displaying them
df.duplicated().sum()
```

```python
#displaying of duplicate values
df[df.duplicated()]
```

```python
#dropping duplicate values
df.drop_duplicates(inplace=True)
```

```python
#Changing the data type of columns to object from float
df['id'] = df['id'].astype(object)
df['host_id'] = df['host_id'].astype(object)
```

```python
#finding all the prices which are above 2000
(df['price']>=2000).sum()
```

```python
#keeping only rows where price is less than 2000
df = df[df['price'] < 2000] 
```

```python
#boxplot for finding outliers
sb.boxplot(data=df,x='price')
```

```python
#Histogram for visualizing Price Distribution
plt.figure(figsize=(15,5))
sb.histplot(data=df,x='price')
plt.ylabel("Frequency")
plt.title("Price Distribution")
```

```python
#Histogram for Availability of properties across the year
plt.figure(figsize=(15,5))
sb.histplot(data=df,x='availability_365')
plt.ylabel("No. of Property")
plt.xlabel("Days")
plt.title("Availability of properties across the year")
```

```python
#Average price per bed of property by neighbourhood group
df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

```python
#new column 'price per bed' = price/bed
df['price per bed'] = df['price']/df['beds']
```

```python
#price of different type of rooms based on neighbourhood
sb.barplot(data=df,x='neighbourhood_group', y='price',hue='room_type')
```
